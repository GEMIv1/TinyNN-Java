```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(r'D:\projects\demo\resources\heart_disease.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
num_cols = data.select_dtypes(include=['number']).columns
cat_cols = data.select_dtypes(exclude=['number']).columns

data[num_cols] = data[num_cols].fillna(data[num_cols].mean())
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
data.isna().sum()
```

```python
data.describe(include='O')
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import pandas as pd

target_col = "Heart Disease Status"

label_encoder = LabelEncoder()
data[target_col] = label_encoder.fit_transform(data[target_col])

cat_cols = data.select_dtypes(include="object").columns

encoder = OneHotEncoder(drop='first', sparse_output=False)

encoded_data = encoder.fit_transform(data[cat_cols])

encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(cat_cols),
    index=data.index
)

data2 = pd.concat(
    [
        data.drop(columns=cat_cols),
        encoded_df
    ],
    axis=1
)
```

```python
data2.info()
```

```python
from sklearn.preprocessing import StandardScaler

target_col = "Heart Disease Status"

scaler = StandardScaler()

feature_cols = data2.columns.drop(target_col)

data2[feature_cols] = scaler.fit_transform(data2[feature_cols])
```

```python
data2.describe()
```

```python
class_counts = data2[target_col].value_counts()
print(f"Original class distribution:\n{class_counts}\n")
    
minority_class = class_counts.idxmin()
majority_class = class_counts.idxmax()
minority_count = class_counts.min()
    
print(f"Minority class: {minority_class} ({minority_count} samples)")
print(f"Majority class: {majority_class} ({class_counts.max()} samples)")
print(f"\nUndersampling majority class to {minority_count} samples...\n")
    
df_minority = data2[data2[target_col] == minority_class]
df_majority = data2[data2[target_col] == majority_class]
    
df_majority_undersampled = df_majority.sample(n=minority_count, random_state=42)
    
df_balanced = pd.concat([df_minority, df_majority_undersampled])
    
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
    
final_counts = df_balanced[target_col].value_counts()
print(f"Final class distribution:\n{final_counts}")
    
```

```python
df_balanced.to_csv("processed_data.csv", index=False)
```
